## 1. NumPy
NumPy is one of the fundamental packages for scientific computing in the browser. 
* Multimensional arrays
* High-level mathematical functions such as linear algebra operations
* Important, because the NumPy array is the fundamental data structure in scikit-learn (which will be mainly used for machine learning algorithms)
* Data will have to be converted to a NumPy array
* Core functionality: ndarray class, multidimensinal (n-dimensional) array
* All elements in the array must be of the same data type (e.g. int, str, float...)

```python
import warnings
warnings.filterwarnings("ignore")

# Import the entire NumPy module
import numpy as np
```

One commonly needed routine is accessing single rows or columns of an array. You can do this by combining indexing and slicing, using an empty slice marked by a single colon (:):

### 6) NumPy's arithmetic operators

```python
x = np.array([1,2,3,4,5,6])
print (x)
```

```python
print (x+3)
print (x-10)
print (x*3)
```

```python
# Absolute, log, exp values
# The inverse of the exponentials, the logarithms, are also available. 
# The basic np.log gives the natural logarithm; if you prefer to compute the base-2 logarithm or 
# the base-10 logarithm, these are available as well:

x = np.array([-10,5,-6,6,5,10])
print ("Original array: ",x)
print ("Absolute values: ", np.abs(x))
print ("Natural logarithm values: ", np.log(x))
print ("Base-10 logarithm values: ", np.log10(x))
print ("Exponential values: ", np.exp(x))
print ("Exponential values: ", np.round(np.exp(x),2))
```

### 7) Calling built-in functions

# SECTION 2: Exploratory data analysis and data viz

```python
# Load required modules and libraries

import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### 1) Load datasets
* 'waste.csv' is solid waste generation data (tons)
* 'features.csv' is neighborhood characeristics data
* There is common column (key column), 'subSection', which is a DSNY administrative district

```python
df1 = pd.read_csv('../data/waste.csv') # solid waste generation in NYC (DSNY district level)
df2 = pd.read_csv('../data/features.csv') # neighborhood characteristics in NYC (DNSY district level)
```

```python
df1.head()
```

```python
df2.head(2)
```

### 2) Check missing values
* This is a really important process to avaoid any wrong results in statistics
* More information about missing data, see 'Python Data Science Handbook, O'Reilly' (p.119)

```python
df1.isnull().sum()
```

* Let's compare mean values of the column 'avgWeekPaper' based on two different methods - 1) not using .mean() function and 2) using .mean() funciton

```python
m1 = df1['avgWeekPaper'].sum() / len(df1['avgWeekPaper'])
m2 = df1['avgWeekPaper'].mean()
print (m1)
print (m2)
```

* This is because Pandas calling functions do not take into account Nulls in the dataframe while doing statistical processing.
* However, we need to consider Nulls because in terms of waste generation or socioeconomic characteristics, Null is 0 (zero). 
* Let's replace Null value to numerical value 0 (zero).

```python
df1['avgWeekPaper'] = df1['avgWeekPaper'].replace(np.nan, 0)
df1['avgWeekMGP'] = df1['avgWeekMGP'].replace(np.nan, 0)
```

```python
df1.isnull().sum()
```

```python
m1 = df1['avgWeekPaper'].sum() / len(df1['avgWeekPaper'])
m2 = df1['avgWeekPaper'].mean()
print (m1)
print (m2)
```

### 3) Create a new column based on the existing columns
* Let's create a new column of total solid waste generation
* Total solid waste generaiton = refuse + paper + metal, glass, and plastic

```python
df1['avgWeekTotal'] = df1['avgWeekRefuse'] + df1['avgWeekPaper'] + df1['avgWeekMGP']
```

```python
df1.head(2)
```

### 4) Join two datasets

* We would like to join two tables (df1 and df2) based on 'subSection' column
* If you want to learn more about mege, join, or concatenate functions of Pandas, here is a useful link (https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html) 
* You can also refer the book 'Python Data Science Handbook, O'Reilly' (p.146)

```python
df = pd.merge(df1, df2, on='subSection', how='left')
```

```python
print (len(df1))
print (len(df2))
print (len(df))
```

### 6) Histogram

```python
#Plot the distribution of refuse waste generation

fig = plt.figure(figsize=(12,8)) # define your figure


df.avgWeekTotal.hist(bins=30, color='grey',edgecolor='black', linewidth=0.5, alpha=0.5) # histogram from pandas dataframe

plt.grid(b=False) # If you make this True, then the gird will be shown

#Plot vertical line - mean
plt.axvline(df.avgWeekTotal.mean(), color='grey', linewidth=2.0, linestyle='--', 
           label="Total waste (Mean)")

#Plot vertical line - median
plt.axvline(df.avgWeekTotal.median(), color='grey', linewidth=1.0, linestyle='--', 
           label="Total waste (Median)")


#std
#Plot vertical line - median
plt.axvline(df.avgWeekTotal.mean()+3*df.avgWeekTotal.std(), color='red', linewidth=1.0, linestyle='--', 
           label="Total waste (3std)")
plt.axvline(df.avgWeekTotal.mean()-3*df.avgWeekTotal.std(), color='red', linewidth=1.0, linestyle='--')

#Adjust labeles (x axis and y axis)
plt.ylabel('Number of DSNY districts', fontsize=13)
plt.xlabel('Average weekly waste collection (tons)', fontsize=13)
#plt.xlim(0,100)

#Legend, title & export
plt.title('Distribution of total waste generation in New York City', fontsize=17)
plt.legend(fontsize=12)
```

### 11) Correlation plot - correlation colormap
* Instead of a correlation table, a correlation colormap is beneficial to understand overall patterns among variables visually very quickly.

```python
import seaborn as sns # for styling

corr = df.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True


# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(10, 8))

# plot the heatmap and annotation on it
sns.heatmap(corr, 
            cmap = 'bwr', 
            square = True,
            mask=mask,
            linewidth=.5,
            alpha=0.7
            )
```

### Choropleth map
* Definition: A choropleth map is a type of thematic map in which areas are shaded or patterned in proportion to a statistical variable that represents an aggregate summary of a geographic characteristic within each area, such as population density or per-capita income.

```python
fig, ax = plt.subplots(1, figsize=(10,10))

district_waste.plot(ax=ax,
                    column='avgWeekTotal', 
                    edgecolor='k',
                    linewidth=0.3,
                    cmap='Greens',
                    k = 5,
                    scheme='quantiles', 
                    legend=False, 
                   )


vmin, vmax = df1['avgWeekTotal'].min(), df1['avgWeekTotal'].max()
# Create colorbar as a legend
sm = plt.cm.ScalarMappable(cmap='Greens', norm=plt.Normalize(vmin=vmin, vmax=vmax))
# empty array for the data range
sm._A = []
# add the colorbar to the figure
cbar = fig.colorbar(sm, fraction=0.025)

ax.set_title('Spatial pattern of weekly solid waste generation in NYC (tons)', fontsize=15)

plt.axis('off')
plt.show()
```
