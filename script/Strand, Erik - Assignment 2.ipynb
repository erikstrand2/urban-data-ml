### 2) Check missing values
* This is a really important process to avaoid any wrong results in statistics
* More information about missing data, see 'Python Data Science Handbook, O'Reilly' (p.119)

```python
df1.isnull().sum()
```

* Let's compare mean values of the column 'avgWeekPaper' based on two different methods - 1) not using .mean() function and 2) using .mean() funciton

```python
m1 = df1['avgWeekPaper'].mean()
print (m1)
```

```python
m2 = df1['avgWeekPaper'].sum() / len(df1['avgWeekPaper'])
print (m2)
```

* This is because Pandas calling functions do not take into account Nulls in the dataframe while doing statistical processing.
* However, we need to consider Nulls because in terms of waste generation or socioeconomic characteristics, Null is 0 (zero). 
* Let's replace Null value to numerical value 0 (zero).

```python
df1['avgWeekPaper'] = df1['avgWeekPaper'].replace(np.nan, 0)
df1['avgWeekMGP'] = df1['avgWeekMGP'].replace(np.nan, 0)
```

```python
df1.isnull().sum()
```

```python
m1 = df1['avgWeekPaper'].mean()
print (m1)
m2 = df1['avgWeekPaper'].sum() / len(df1['avgWeekPaper'])
print (m2)
```

* Total solid waste generaiton = refuse + paper + metal, glass, and plastic

```python
df1['avgWeekTotal'] = df1['avgWeekRefuse'] + df1['avgWeekPaper'] + df1['avgWeekMGP']
```

* Definition: A choropleth map is a type of thematic map in which areas are shaded or patterned in proportion to a statistical variable that represents an aggregate summary of a geographic characteristic within each area, such as population density or per-capita income.

```python
fig, ax = plt.subplots(1, figsize=(10,10))

district_waste.plot(ax=ax,
                    column='avgWeekTotal', 
                    edgecolor='k',
                    linewidth=0.3,
                    cmap='Greens',
                    k = 5,
                    scheme='quantiles', 
                    legend=False, 
                   )


vmin, vmax = df1['avgWeekTotal'].min(), df1['avgWeekTotal'].max()
# Create colorbar as a legend
sm = plt.cm.ScalarMappable(cmap='Greens', norm=plt.Normalize(vmin=vmin, vmax=vmax))
# empty array for the data range
sm._A = []
# add the colorbar to the figure
cbar = fig.colorbar(sm, fraction=0.025)

ax.set_title('Spatial pattern of weekly solid waste generation in NYC (tons)', fontsize=15)

plt.axis('off')
plt.show()
```

## 2. For the df_waste dataframe, create new columns named 'recycling_rate'.
    Recycling rate (also known as diversion rate) is recycling generation divided by total waste generation. Recycling materials are represented as 'avgWeekPaper' and 'avgWeekMGP' (Metal, glass, and plastic). In order to calculate total solid waste generation, you need to sum all waste streams (refuse, paper, and MGP).

```python
df_waste["recycling_rate"] = (df_waste["avgWeekPaper"] + df_waste["avgWeekMGP"]) / (df_waste["avgWeekPaper"] + df_waste["avgWeekMGP"] + df_waste["avgWeekRefuse"])
df_waste.head()
```

## 3. Use either a histogram or a box plot to describe a citywide pattern of recycling rates.

```python
fig = plt.figure(figsize=(12,8))

df_waste.recycling_rate.hist(bins=30, color='grey',edgecolor='black', linewidth=0.5, alpha=0.5) # histogram from pandas dataframe

plt.grid(b=False) # If you make this True, then the gird will be shown

#Plot vertical line - mean
plt.axvline(df_waste.recycling_rate.mean(), color='grey', linewidth=2.0, linestyle='--', 
           label="Total waste (Mean)")

#Plot vertical line - median
plt.axvline(df_waste.recycling_rate.median(), color='grey', linewidth=1.0, linestyle='--', 
           label="Total waste (Median)")


#std
#Plot vertical line - median
plt.axvline(df_waste.recycling_rate.mean()+3*df_waste.recycling_rate.std(), color='red', linewidth=1.0, linestyle='--', 
           label="Total waste (3std)")
plt.axvline(df_waste.recycling_rate.mean()-3*df_waste.recycling_rate.std(), color='red', linewidth=1.0, linestyle='--')

#Adjust labeles (x axis and y axis)
plt.ylabel('Number of DSNY districts', fontsize=13)
plt.xlabel('Proportion of weekly waste recycled', fontsize=13)
#plt.xlim(0,100)

#Legend, title & export
plt.title('Recycling rates in New York City', fontsize=17)
plt.legend(fontsize=12)
```
